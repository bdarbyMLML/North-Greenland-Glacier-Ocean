```python
time_ar = []
def iter_number_to_date(iter_number,seconds_per_iter=60):

    total_seconds = iter_number*seconds_per_iter
    date = datetime(1992,1,1) + dt.timedelta(seconds=total_seconds)
    # print(date)
    return(date)

for t in np.array(total_melt.dropna(dim='iterations',how='all').iterations):
    time_ar.append(iter_number_to_date(t,seconds_per_iter=30))
```

```python
time_ar
total_melt = total_melt.rename({'iterations':'time'})
total_melt['time']= time_ar

total_melt.plot()
```
